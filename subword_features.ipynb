```python
def remove_special_characters(words):
    cleaned_text_list = []  # Using a set to avoid duplicates

    for word in words:
      if word in words_to_ignore:
        cleaned_text_list.append((word, '-99'))
      else:
        cleaned_text = re.sub(r'[^a-zA-Z0-9\s]', '', word)
        removed_chars = re.findall(r'[^a-zA-Z0-9\s]', word)
        cleaned_text_list.append((cleaned_text, removed_chars))

        #cleaned_text_list.append(cleaned_text)
        #print(cleaned_text)

    return cleaned_text_list
```

```python
# Example usage
#words = ["hello", "world!", "123", "#$%"]
#special_characters = remove_special_characters(words)
#print("Special characters:", remove_special_characters)
cleaned_text_list = remove_special_characters(words)
cleaned_text_only = [cleaned_text for cleaned_text, special_char in cleaned_text_list]
special_char_only = [special_char for cleaned_text, special_char in cleaned_text_list]
print(cleaned_text_only)
print(special_char_only)
#print("Cleaned text only:", cleaned_text_only)
```

#### Tokenization


```python
from transformers import GPT2Tokenizer, GPT2Model
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
```

```python
def tokenize_words_into_subwords(words):
  subword_list = []
  makeup_word_list = []
  for word in words:
    if word in words_to_ignore:
      subword_list.append(word)
      makeup_word_list.append(word)
    else:
      tokens = tokenizer.tokenize(word)
      makeup_word = tokenizer.convert_tokens_to_string(tokens)
      #print(makeup_word)
      subword_list.append(tokens)
      makeup_word_list.append(makeup_word)
  return subword_list, makeup_word_list

```

```python
words = words
subwords, makeup_words = tokenize_words_into_subwords(words)
print("subwords:", subwords)
print('makeup words:', makeup_words)
```

```python
cleaned_subwords, cleaned_makeup_words = tokenize_words_into_subwords(cleaned_text_only)
print("subwords:", cleaned_subwords)
print('makeup words:', cleaned_makeup_words)
```

```python
# number of sub-words  and the length of each sub-word

def calculate_subword_info(subwords):
  num_subwords = []
  subword_lengths = []
  subword_infos = []
  for sw in subwords:
    if sw in words_to_ignore:
      num_subwords.append(-99)
      subword_lengths.append(-99)
    else:
      num_subwords.append(len(sw))
      subword_length = [len(sub) for sub in sw]
      subword_lengths.append(subword_length)
    #print(swlen)

  for i, sw in enumerate(subwords, 1):           # Enumerate over subwords
    if sw in words_to_ignore:
      subword_infos.append(-99)
    else:
      subword_info = [f"{i}-{j}" for j in range(1, len(sw) + 1)]  # Creating subword info for each subword
      subword_infos.append(subword_info)

  return num_subwords, subword_lengths, subword_infos
```

```python
#subwords = [['F', 'ertility'], ['Emb', 'ry', 'ology']]
num_subwords, subword_lengths, subword_infos = calculate_subword_info(subwords)

print("Number of subwords:", num_subwords)
print("Length of each subword:", subword_lengths)
print("Subword info:", subword_infos)
```
